```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1= pd.read_csv(r"..\olympics\Olympic paris\athletes.csv", index_col = "country_code")
```

```python
df2 = df1[['country','code','name']]
```

```python
df2.head(2)
```

```python
df3 = df2.groupby(df2.index).size()
```

```python
total_number_of_athletes = df3.sum()
```

```python
df3 = df3.to_frame(name='total_number_of_athletes_for_a_country')
```

```python
df3['total_number_of_athletes'] = total_number_of_athletes
```

```python
df3['total_athletes_per_country/total_athletes'] = round(df3['total_number_of_athletes_for_a_country']/df3['total_number_of_athletes'],2)
```

```python
df4 = df3.sort_values(by='total_athletes_per_country/total_athletes', ascending=False)
```

```python
df5 = df4[(df4['total_athletes_per_country/total_athletes']>=0.01)]
```

```python
df5.count()
```

```python
df5.head(5)
```

```python
dff1= pd.read_csv(r"C:\Users\lasra\Desktop\olympics\Olympic paris\medals_total.csv", index_col = "country_code")
```

```python
merged_df = dff1.merge(df5, left_index=True, right_index=True, how='inner')  # or 'outer', 'left', 'right'
```

```python
merged_df
```

```python
merged_df['total_country_medal_per_athlete'] = \
         round(merged_df['Total']/merged_df['total_number_of_athletes_for_a_country'],2)
```

```python
merged_df['total_country_gold_medal_per_athlete'] = \
         round(merged_df['Gold Medal']/merged_df['total_number_of_athletes_for_a_country'],2)
```

```python
merged_df
```

```python
df_final = merged_df[['total_athletes_per_country/total_athletes','total_country_medal_per_athlete','total_country_gold_medal_per_athlete']]
```

```python
df_export = df_final.head(10)
```

```python
df_export
```

```python
import matplotlib.pyplot as plt

# Sorting data for 'total_country_medal_per_athlete'
df_medal = df_export.sort_values(by='total_athletes_per_country/total_athletes', ascending=False)

# Plotting the horizontal bar chart
fig, ax1 = plt.subplots(figsize=(10, 6))

# Horizontal bar chart for 'total_country_medal_per_athlete'
ax1.barh(df_medal.index, df_medal['total_athletes_per_country/total_athletes'], color=plt.cm.tab20.colors[:len(df_medal)])
for index, value in enumerate(df_medal['total_athletes_per_country/total_athletes']):
    ax1.text(value, index, f'{value:.2f}', va='center')

ax1.set_title('total_athletes_per_country/total_athletes')
ax1.set_xlabel('total_athletes_per_country/total_athletes')
ax1.invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
# Sorting data for 'total_country_medal_per_athlete'
df_medal = df_export.sort_values(by='total_country_medal_per_athlete', ascending=False)

# Plotting the horizontal bar chart
fig, ax1 = plt.subplots(figsize=(10, 6))

# Horizontal bar chart for 'total_country_medal_per_athlete'
ax1.barh(df_medal.index, df_medal['total_country_medal_per_athlete'], color=plt.cm.tab20.colors[:len(df_medal)])
for index, value in enumerate(df_medal['total_country_medal_per_athlete']):
    ax1.text(value, index, f'{value:.2f}', va='center')

ax1.set_title('Total Medals per Athlete by Country')
ax1.set_xlabel('Total Medals per Athlete')
ax1.invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sorting data for 'total_country_gold_medal_per_athlete'
df_gold_medal = df_export.sort_values(by='total_country_gold_medal_per_athlete', ascending=False)

# Plotting the horizontal bar chart
fig, ax2 = plt.subplots(figsize=(10, 6))

# Horizontal bar chart for 'total_country_gold_medal_per_athlete'
ax2.barh(df_gold_medal.index, df_gold_medal['total_country_gold_medal_per_athlete'], color=plt.cm.tab20.colors[:len(df_gold_medal)])
for index, value in enumerate(df_gold_medal['total_country_gold_medal_per_athlete']):
    ax2.text(value, index, f'{value:.2f}', va='center')

ax2.set_title('Total Gold Medals per Athlete by Country')
ax2.set_xlabel('Total Gold Medals per Athlete')
ax2.invert_yaxis()

plt.tight_layout()
plt.show()
```
