```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'..\olympics\Olympic paris\medals_total.csv')
```

```python
# Calculate the gold medal share for each country
data['Gold Medal Share'] = round(data['Gold Medal'] / data['Total'],2)

# Display the updated dataset with the gold medal share
data
```

```python
# Calculate the summation of gold and silver medals for each country
data['Gold + Silver'] = data['Gold Medal'] + data['Silver Medal']

# Calculate the ratio of the summation of gold and silver over the total number of medals
data['Gold + Silver Medal Share'] = round(data['Gold + Silver'] / data['Total'],2)

# Display the updated dataset with the new ratio
data
```

```python
# Calculate the share of gold medals from the summation of gold and silver medals
data['Gold from Gold + Silver Share'] = round(data['Gold Medal'] / data['Gold + Silver'],2)

# Display the updated dataset with the gold share from gold and silver
data
```

```python
data2 = data.sort_values('Gold Medal', ascending = False).head(5)
```

```python
data2
```

```python
data3 = data2 [['country_code','Gold Medal Share','Gold + Silver Medal Share','Gold from Gold + Silver Share']]
```

```python
data3_sorted = data3.sort_values('Gold from Gold + Silver Share', ascending=False).head()
```

```python
# Set the country codes as the index
data3_sorted.set_index('country_code', inplace=True)

# Create the heatmap using the sorted data
plt.figure(figsize=(10, 6))
sns.heatmap(data3_sorted, annot=True, cmap='coolwarm', cbar=True, linewidths=.5)

plt.title('Heatmap of Medal Shares for Top Countries (Data Source: data3)')
plt.show()
```
